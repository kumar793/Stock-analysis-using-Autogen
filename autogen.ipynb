```python
import os 
from dotenv import load_dotenv
load_dotenv()
os.environ["GEMINI_API_KEY"] = os.getenv("GEMINI_API_KEY")
api = os.getenv("GEMINI_API_KEY")
```

```python
from autogen import ConversableAgent
config_list  = [
    {
        "model": "gemini-2.0-flash",
        "api_key": api,
        "api_type": "google"
    }
]
```

```python
llm_config = {
    "config_list" :config_list,
    "cache":None
}
```

```python
agent  = ConversableAgent(
    name = "Chatbot",
    llm_config=llm_config,
    system_message="""You are the Financial advisor who advises
    stocks to invest in a most reputed financial firm. Suggest the stocks with 
    fundamental analysis, technical analysis and its target and stop loss . """,
    human_input_mode="NEVER"
)
```

```python
reply = agent.generate_reply([{"content":"give details on tata steel stock","role":"user"}])
```

# 2- Turn conversation

```python
agent1 = ConversableAgent(
    name = "Kumar",
    llm_config=llm_config,
    system_message="""
your name is kumar, You are a telugu poet in a 2 poet telugu poetry show."
"""
)

agent2 = ConversableAgent(
    name = "Kumar",
    llm_config=llm_config,
    system_message="""
your name is ram, You are a telugu poet in a 2 poet telugu poetry show."
"""
)
```

```python
reply = agent1.initiate_chat(
    recipient=agent2,
    message="tell me all the complete poems  in order in gajendra mokshanam where vishnu starts from vaikuntam to climax in telugu",
    max_turns=10,
    summary_method="reflection_with_llm",
    summary_prompt = "summerize the conversation"
)
```

```python
from pprint import pprint as p
p(reply.chat_history)
```

```python
p(reply.cost)
```

```python
reply.summary
```
